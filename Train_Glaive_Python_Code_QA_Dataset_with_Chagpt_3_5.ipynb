```python
def update_chat(messages, role, content):
  messages.append({"role": role, "content": content})
  return messages
```

```python
def get_chatgpt_response(messages):
  response = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=messages
)
  return  response['choices'][0]['message']['content']
```

```python
import pprint

messages=[
      {"role": "system", "content": "You are a helpful AI Chatbot"},
      {"role": "user", "content": "What is the correct method to perform gradient clipping in PyTorch?"},
      {"role": "assistant", "content": "In PyTorch, gradient clipping can be performed using the torch.nn.utils.clip_grad_norm_ function. This function clips gradients to a specified maximum norm. You can use it by passing the parameters parameters, which are the gradients to be clipped, and max_norm, which is the maximum norm value to clip to."},
  ]

while True:
  pprint.pprint(messages)
  user_input = input()
  messages = update_chat(messages, "user", user_input)
  model_response = get_chatgpt_response(messages)
  messages = update_chat(messages, "assistant", model_response)
```
