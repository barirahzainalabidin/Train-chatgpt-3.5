```python
def update_chat(messages, role, content):
  messages.append({"role": role, "content": content})
  return messages
```

```python
def get_chatgpt_response(messages):
  response = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=messages
)
  return  response['choices'][0]['message']['content']
```
